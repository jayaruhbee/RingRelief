TQ - Tinnitus Questionnaire: The TQ is a commonly used self-report questionnaire that assesses the severity and impact of tinnitus on an individual's life. It includes questions about the loudness and distress associated with tinnitus and is used to measure the subjective experience of tinnitus.

THI - Tinnitus Handicap Inventory: The THI is another self-report questionnaire designed to assess the degree of handicap experienced by individuals with tinnitus. It measures the emotional and functional impact of tinnitus on a person's daily life.

LMT - Loudness Matching Test: The LMT is a test used to measure the perceived loudness of an individual's tinnitus. It involves matching the loudness of the tinnitus to external sounds or tones.

```python
import matplotlib.pyplot as plt

# Filter columns for "before" and "after" measurements
before_columns = ['LMTpre', 'thipre', 'tqpre', 'emotionaldpre', 'cognetivedpre', 'ecpre', 'intrusivenesspre', 'APDpre', 'sleepdispre', 'somaticcompre']
after_columns = ['LMTpost', 'thipost', 'tqpost', 'emotionaldpost', 'cognetivedpost', 'ecpost', 'inrusivenesspost', 'APDpost', 'sleepdispost', 'somaticcopost']

# Calculate summary statistics for "before" and "after" results
before_summary = df[before_columns].mean()
after_summary = df[after_columns].mean()

# Create a bar plot for "before" and "after" results
plt.figure(figsize=(12, 6))
plt.bar(before_summary.index, before_summary, color='blue', alpha=0.7, label='Before Intervention')
plt.bar(after_summary.index, after_summary, color='green', alpha=0.7, label='After Intervention')
plt.xticks(rotation=45)
plt.xlabel('Measurements')
plt.ylabel('Mean Value')
plt.title('Summary of Measurements Before and After Intervention')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Define the measurements
measurements = before_columns

# Create a DataFrame to store the averages
averages_df = pd.DataFrame({'Measurements': measurements, 'Before Intervention': before_summary.values, 'After Intervention': after_summary.values})

# Save the DataFrame to a CSV file
averages_df.to_csv("/content/drive/MyDrive/Colab Notebooks/averages.csv", index=False)
```
